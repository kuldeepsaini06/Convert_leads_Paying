```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
def draw_roc(actual,prob):
    fpr,tpr,thresholds=roc_curve(actual,prob,drop_intermediate=False)
    auc_score=roc_auc_score(actual,prob)
    plt.figure(figsize=(5,5))
    plt.plot(fpr,tpr,label='Roc Curve (area=%0.2f)' % auc_score)
    plt.plot([0,1],[0,1],'k--')
    plt.xlim(0.0,1.0)
    plt.ylim(0.0,1.05)
    plt.legend(loc='lower right')
    plt.show()
    
    return None
```

```python
fpr,tpr,thresholds=roc_curve(y_train_pred_final.Converted,y_train_pred_final.Conversion_prob)
```

```python
draw_roc(y_train_pred_final.Converted,y_train_pred_final.Conversion_prob)
```

```python
numbers= [float(x)/10 for x in range(10)]
```

```python
for i in numbers:
    y_train_pred_final[i]=y_train_pred_final.Conversion_prob.map(lambda x: 1 if x>i else 0)
```

```python
y_train_pred_final.head(10)
```

```python
cutoff_df=pd.DataFrame(columns=['prob','accuracy','sensitivity','specificity'])
```

```python
numbers
```

```python
for i in numbers:
    cm1=confusion_matrix(y_train_pred_final['Converted'],y_train_pred_final[i])
    total1= sum(cm1)
    accuracy=(cm1[0,0]+cm1[1,1])/total1
    specificity=(cm1[0,0])/(cm1[0,0]+cm1[0,1])
    sensitivity= (cm1[1,1])/(cm1[1,1]+cm1[1,0])
    cutoff_df.loc[i]=[i,acuracy,sensitivity,specificity]
```
